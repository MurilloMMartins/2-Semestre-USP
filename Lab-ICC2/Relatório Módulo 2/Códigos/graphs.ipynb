```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data2.csv")
df
```

```python
x = list(df["TAM"].unique())
by_size = [df.loc[df["TAM"] == t] for t in x]
y_q = [np.mean(x["QUICK_SORT"]) for x in by_size]
y_h = [np.mean(x["HEAP_SORT"]) for x in by_size]


e_q = [np.std(x["QUICK_SORT"]) for x in by_size]
e_h = [np.std(x["HEAP_SORT"]) for x in by_size]
```

```python
plt.style.use("seaborn")
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio - QuickSort [segundos]')

plt.errorbar(x, y_q, yerr=e_q, fmt='o--', ecolor='r', capsize=3)
```

```python
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio - HeapSort [segundos]')

plt.errorbar(x, y_h, yerr=e_h, fmt='o--', ecolor='r', capsize=3)
```

```python
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [segundos]')

plt.plot(x,y_q, 'o--', label='QuickSort')
plt.plot(x,y_h, 'o--', label='HeapSort')

plt.legend()
```

```python
dt = pd.read_csv("data1.csv")
x = list(dt["TAM"].unique())
by_size = [dt.loc[df["TAM"] == t] for t in x]
y_b = [np.mean(x["BUBBLE_SORT"]) for x in by_size]
y_i = [np.mean(x["INSERTION_SORT"]) for x in by_size]
y_m = [np.mean(x["MERGE_SORT"]) for x in by_size]

plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [segundos]')

plt.plot(x,y_b, 'o--', label='BubbleSort')
plt.plot(x,y_i, 'o--', label='InsertionSort')
plt.plot(x,y_m, 'o--', label='MergeSort')
plt.plot(x,y_q, 'o--', label='QuickSort')
plt.plot(x,y_h, 'o--', label='HeapSort')


ax = plt.gca()
ax.set_ylim([0, 0.025])
plt.legend()
```
