```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df
```

```python
x = list(df["TAM"].unique())
by_size = [df.loc[df["TAM"] == t] for t in x]
y_b = [np.mean(x["BUBBLE_SORT"]) for x in by_size]
y_i = [np.mean(x["INSERTION_SORT"]) for x in by_size]
y_m = [np.mean(x["MERGE_SORT"]) for x in by_size]


e_b = [np.std(x["INSERTION_SORT"]) for x in by_size]
e_i = [np.std(x["INSERTION_SORT"]) for x in by_size]
e_m = [np.std(x["INSERTION_SORT"]) for x in by_size]
```

```python
plt.style.use("seaborn")
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio - Bubble Sort [s]')

plt.errorbar(x, y_b, yerr=e_b, fmt='o--', ecolor='r', capsize=3)
```

```python
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio - Insertion Sort [s]')

plt.errorbar(x, y_i, yerr=e_i, fmt='o--', ecolor='r', capsize=3)
```

```python
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio - Merge Sort [s]')

plt.errorbar(x, y_m, yerr=e_m, fmt='o--', ecolor='r', capsize=3)
```

```python
plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execução médio [s]')

plt.plot(x,y_b, 'o--', label='Bubble')
plt.plot(x,y_i, 'o--', label='Insertion')
plt.plot(x,y_m, 'o--', label='Merge')

plt.legend()
```
